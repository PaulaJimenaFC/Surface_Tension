### Grafica del material suplementario

```python
import matplotlib.pyplot as plt
import glob

# Obtener la lista de rutas de todas las imgenes PNG en la carpeta
rutas_imagenes = glob.glob('*.png')

# Crear una figura con una grilla de 4 filas y 2 columnas
fig = plt.figure(figsize=(10, 20), dpi=80)

# Iterar sobre las rutas de las imgenes y agregarlas a la figura en las posiciones correctas
for i, ruta_imagen in enumerate(rutas_imagenes):
    imagen = plt.imread(ruta_imagen)
    ax = fig.add_subplot(4, 2, i+1)
    ax.imshow(imagen)
    ax.axis('off')

# Ajustar los espacios entre los subplots
fig.tight_layout()

# Guardar la figura en alta definicin
fig.savefig('grilla_imagenes.png', dpi=300)

# Mostrar la figura
plt.show()
```
